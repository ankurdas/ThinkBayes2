```python
from thinkbayes2 import Pmf
import thinkplot
```

```python
d6 = Pmf(range(1,7))
two = d6+Pmf(range(1,7))
three = two+Pmf(range(1,7))
four = three+Pmf(range(1,7))
five = four+Pmf(range(1,7))
thinkplot.Pdfs((d6,two,three,four,five))


```

```python
print(three.Mean())
print(three.Var())
print(three.Var()**.5)
```

```python
print(three.ProbGreater(15))
```

```python
print(three.Random())
```
