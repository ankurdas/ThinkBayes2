```python
"""This file contains code for use with "Think Bayes",
by Allen B. Downey, available from greenteapress.com


Copyright 2014 A
llen B. Downey
License: GNU GPLv3 http://www.gnu.org/licenses/gpl.html """

from __future__ import print_function, division

import numpy
import thinkbayes2
import thinkplot
import math
import scipy

class Hyreplace(thinkbayes2.Suite):
    """Represents hypotheses about the state of the electorate."""

    def Likelihood(self, data, hypo):
        """Computes the likelihood of the data under the hypothesis.
        hypo:
        data:
        """
        
        Nhypo = hypo
        common, tagged, caught = data
    
        like = EvalHyperGeometricPmf(common, Nhypo, tagged, caught)
        if math.isnan(like):
            like = 0
        return like
    
class Hyrax(thinkbayes2.Suite):
    """Represents hypotheses about the state of the electorate."""

    def Likelihood(self, data, hypo):
        """Computes the likelihood of the data under the hypothesis.
        hypo:
        data:
        """
        
        Nhypo = hypo*1.0
        common, initial, caught = data
        p = initial/Nhypo
    
        like = thinkbayes2.EvalBinomialPmf(common, caught, p)
        if math.isnan(like):
            like = 0.0
        return like

class Hyreplace(thinkbayes2.Suite):
    """Represents hypotheses about the state of the electorate."""

    def Likelihood(self, data, hypo):
        """Computes the likelihood of the data under the hypothesis.
        hypo:
        data:
        """
        
        Nhypo = hypo
        common, tagged, caught = data
    
        like = EvalHyperGeometricPmf(common, Nhypo, tagged, caught)
        if math.isnan(like):
            like = 0
        return like
 
def EvalHyperGeometricPmf(k, M, n, N):
    """Evaluates the hypergeometric pmf.

    Returns the probabily of k successes in N trials with n tagged 
    items out of M total.
    """
    return scipy.stats.hypergeom.pmf(k, M, n, N)

def main():
    hypos = numpy.linspace(0, 500, 501)
    suite1 = Hyreplace(hypos)
    suite2 = Hyrax(hypos)

    thinkplot.Pdf(suite1, label='prior')

    data = 2, 5, 5
    suite1.Update(data)
    suite2.Update(data)

    thinkplot.Pdf(suite1, label='posterior1')
    thinkplot.Pdf(suite2, label='posterior2')
    
    
    thinkplot.Show()
    
    print(suite1.Mean())
    print(suite1.Percentile(5),suite1.Percentile(95))
    
    print(suite2.Mean())
    print(suite2.Percentile(5),suite2.Percentile(95))

if __name__ == '__main__':
    main()
```
